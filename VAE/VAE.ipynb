# Auto Encoders

### Overview

To talk about Variational AutoEncoders (VAE), we must first udnerstand what an AutoEncoder (AE) is.

In AEs, the objective is to reconstruct a data distribution **_P(X)_**, where **_X_** is the data and **_P(X)_** is the data label probabilitiy distribution.

At the core of the AutoEncoder (AE) lies a three-component architecture: 
1. The encoder
2. The bottleneck ( latent space, often denoted as **_z_** )
3. The decoder

**1. Encoder: Mapping Input Data to Latent Space**

The encoder, denoted as **_Q_(z | x)**, is a neural network parameterized by **ϕ** that takes an input **_x_** (∈ R^n) and maps it to a distribution over a latent variable **_z_** (∈ R^d), where typically d≪n. 

In simple terms the encoder projects the input data into a much smaller space
